```python
import PyPDF2
from sapling import SaplingClient
from pprint import pprint
```

```python
def extract_text_from_pdf(pdf_path):
    pdf_reader = PyPDF2.PdfReader(open(pdf_path, 'rb'))
    text = ""
    for page_num in range(len(pdf_reader.pages)):

        page = pdf_reader.pages[page_num]
        text += page.extract_text()
    return text
```

```python
path = '/kaggle/input/input-paper/t2speech.pdf'
text = extract_text_from_pdf(path)
```

```python
print(text[:400])
```

```python
def split_into_chunks(text, chunk_size=300):
  return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

chunks = split_into_chunks(text)
```

# 1.Demo of AI content detection using Sampling AI 

```python
import json
import requests

def process_pdf(file_path, json_params):
    url = 'https://api.sapling.ai/api/v1/ingest/pdf_to_text'
    files = {
        'file': open(file_path, 'rb'),
        'jsonParams': ('jsonParams', json.dumps(json_params), 'application/json')
    }

    response = requests.post(url, files=files)

    if response.status_code == 200:
        print(response.json())

    else:
        print(f'Failed, status code: {response.status_code}')
        print(response.text)

    return response.json()
```
